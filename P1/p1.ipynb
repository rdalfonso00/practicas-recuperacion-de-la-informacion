```python
filename = "julioverne.txt"
file = open(filename, 'rt',encoding='utf-8-sig')
text = file.read()
file.close()
```

3. Separa el texto por palabras utilizando el espacio como delimitador y escribe las primeras 100 palabras
¿Qué observas en estas palabras?

```python
words = text.split()
print("*****First 100 words*****\n")
print(words[:100])
```

4. Utiliza re de python para eliminar signos de puntuación. Lee en la web como funciona re en python, escribe un breve resumen de lo que entendiste. Ahora prueba las siguientes líneas, asegúrate de entenderlas y hacer cambios si es necesario.

```python
import re
words = re.split(r'\W+',text)
print("*****Words without puctuation with re*****\n")
print(words[:100])
```

```python
import string
import re
print(string.punctuation+"\n")
re_punc = re.compile('[%s]' % re.escape(string.punctuation))
stripped = [re_punc.sub("",w) for w in words]

print("\n*****Stripped words*****\n")                   
print(stripped[:100])
```

5. Convierte a minúsculas el texto, utiliza lower(), lee como funciona este método y escribe para que sirve. Utilízalo y escribe nuevamente las 100 primeras palabras.

Escribe tus observaciones.


método string.lower() -> convertir a minúsculas una cadena
se le aplica a cadenas de texto (se tiene que aplicar antes de que se separen las palabras)


```python
import string
text_lower = text.lower()
words = re.split(r'\W+',text_lower)
re_punc = re.compile('[%s]' % re.escape(string.punctuation))
words_lower = [re_punc.sub("",w) for w in words]
 

print("*****First 100 words lowered*****\n")
print(words_lower[:100])
```

6. Ahora elimina palabras vacías

```python
# Primero, es necesario descargar las stopwords de la red
import nltk
nltk.download('stopwords')
```

```python
# Ahora, ya se pueden importar estas palabras sin mucho problema
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
```

Estas palabras son, en su mayoría, conectores, prefijos, sufijos, entre otros recursos literarios que, en un proceso de recuperación de la información, solamente retrasarían el procesamiento de los datos.

```python
# Para eliminar las palabras vacías, se hace uso de las operaciones de list comprehension de python
words_filtered = [x for x in words_lower if x not in stop_words]
print(words_filtered[:100])
```
