#### Explique brevemente como se utiliza la presión R para comparar SRI
La precisión-R es definida como la proporción de los primeros _R_ documentos recuperados relevantes.
Donde R es el número de documentos relevantes del query actual.
El comparar el promedio de precisiones-R entre 2 SRI es útil en el caso de que se juzquen una larga cantidad de documentos relevantes.

Al calcular la precision-R de los querys aplicados a un par de SRI, podemos hacer dos tipos de análisis para determinar el mejor de los dos.
Sean SRI_A y SRI_B dos sistemas de recuperación de la información. Tenemos:

1. Comparar query a query en cuántos de estos es mejor el SRI_A vs el SRI_B
1. Al computar el promedio de las precisiones-R, podemos determinar el SRI que es mejor que el otro en conjunto.

Primero, importamos los archivos stats.csv de nuestra práctica y el de el equipo al que se le solicitaron


```python
import pandas as pd
df_A = pd.read_csv("assets/stats.csv")
df_A.drop("Unnamed: 0", axis=1, inplace=True)

display(df_A)
```

```python
df_B = pd.read_excel("assets/Datos.xlsx")
df_B.drop("Unnamed: 0", axis=1, inplace=True)
display(df_B)
```

Podemos observar un resumen rápido de las columnas Precision-R con la función display()

```python
display(df_A.describe())
display(df_B.describe())
```

Viendo este resumen rápido, podemos observar que la media de los datos es mayor en el sistema A.

Para generar conclusiones más contundentes, se realizará la tabla recomendada en la descripción de la práctica.

```python
# convirtiendo a np.array
import numpy as np
pres_A = np.array(df_A['Precision-R'])
pres_B = np.array(df_B['Precision-R'])
# calculando la diferencia
dif_A_B = pres_A - pres_B
# rearmando en un nuevo df para comparar:

df_res = pd.DataFrame(np.array([pres_A.T, pres_B.T, dif_A_B.T]), 
                      index=["PR_A(i)", "PR_B(i)", "PR_A(i) - PR_B(i)"]).T

# contar las veces en que PR_A > PR_B
A_count = df_res[df_res['PR_A(i) - PR_B(i)'] > 0 ]['PR_A(i) - PR_B(i)'].count()
B_count = df_res[df_res['PR_A(i) - PR_B(i)'] < 0 ]['PR_A(i) - PR_B(i)'].count()
cero_count = 64 - A_count - B_count

df_res.loc[len(df_res.index)+1] = [np.average(pres_A), np.average(pres_B), np.average(dif_A_B)]
display(df_res)
print(f'A_count: {A_count}')
print(f'B_count: {B_count}')
print(f'cero_count: {cero_count}')
```

Con la tabla anterior, se cumplen los dos puntos a notar en el planteamiento de cómo comparar dos SRI.
Se muestra que el SRI-A tiene una precisión-R promedio de 25.41%, mientras que el B solo 10.66%.

Además, de los 64 queries, se nota que los sistemas son iguales en 25/64 queries, mientras que el A es mejor que el B en 33 ocasiones, siendo el B mejor que el A solo en 6 insntancias.
